```python
from bs4 import BeautifulSoup
import requests

URL = "https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors/page"

rank = []
major = []
degree_type = []
early_career_pay = []
mid_career_pay = []
high_meaning = []

for page in range(34):
    response = requests.get(f"{URL}/{page+1}")
    soup = BeautifulSoup(response.text, "html.parser")
    sal_data = soup.select(selector=".data-table__value")
    for i in range(len(sal_data)):
        if i % 6 == 5:
            high_meaning.append(sal_data[i].text)
        elif i % 6 == 4:
            mid_career_pay.append(sal_data[i].text)
        elif i % 6 == 3:
            early_career_pay.append(sal_data[i].text)
        elif i % 6 == 2:
            degree_type.append(sal_data[i].text)
        elif i % 6 == 1:
            major.append(sal_data[i].text)
        else:
            rank.append(sal_data[i].text)

dic = {"Rank": rank, "Major": major, "Degree Type": degree_type, "Early Career Pay": early_career_pay, "Mid Career Pay": mid_career_pay, "High Meaning": high_meaning}

new_df = pd.DataFrame(dic)
new_df.head()
```

```python
new_df.describe()
```

```python
df_2 = new_df[["Major", "Early Career Pay", "Mid Career Pay"]]
df_2.head()
```

```python
df_2.loc[:, "Early Career Pay"] = df_2.loc[:, "Early Career Pay"].str.replace("$", "")
df_2.loc[:, "Early Career Pay"] = df_2.loc[:, "Early Career Pay"].str.replace(",", "")
df_2.loc[:, "Mid Career Pay"] = df_2.loc[:, "Mid Career Pay"].str.replace("$", "")
df_2.loc[:, "Mid Career Pay"] = df_2.loc[:, "Mid Career Pay"].str.replace(",", "")
df_2.head()
```

```python
df_2.loc[:, "Early Career Pay"] = pd.to_numeric(df_2.loc[:, "Early Career Pay"])
df_2.loc[:, "Mid Career Pay"] = pd.to_numeric(df_2.loc[:, "Mid Career Pay"])
```
